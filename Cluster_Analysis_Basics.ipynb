# Basics of Cluster Analysis

## Import the relevant libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from sklearn.cluster import KMeans
```

## Load the Data

```python
data = pd.read_csv('/content/3.01. Country clusters.csv')
data.head()
```

## Plot the Data

```python
plt.scatter(data['Longitude'], data['Latitude'])
plt.xlim(-180, 180)
plt.ylim(-90, 90)
plt.show()
```

## Select the Features

**DataFrame.iloc(row indices, column indices)** : slices the data frame, given the rows and columns to be kept

```python
x= data.iloc[:, 1:3]
```

```python
x.head()
```

## Clustering

- the value in brackets us K (the number of clusters)

```python
kmeans= KMeans(3)
```

```python
kmeans.fit(x)
```

## Clustering Resutlts

**sklearn.cluster.KMeans.fit_predict(x)**: returns the cluster predictions in an array

```python
identified_clusters= kmeans.fit_predict(x)
identified_clusters
```

```python
data_w_clusters= data.copy()
data_w_clusters['Cluster']= identified_clusters
data_w_clusters
```

```python
plt.scatter(x['Longitude'], x['Latitude'], c= data_w_clusters['Cluster'], cmap= 'rainbow')
plt.xlim(-180, 180)
plt.ylim(-90, 90)
plt.show()
```
