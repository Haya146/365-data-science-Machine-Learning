# Feature selection through Standradization

## Import the libraries


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.linear_model import LinearRegression
```

## Load the data

```python
data = pd.read_csv('1.02. Multiple linear regression.csv')
data.head()
```

```python
data.describe()
```

## create the multiple linear regression

### Declare the dependent and independent variables

```python
x= data[['SAT','Rand 1,2,3']]
y=data['GPA']
```

## Standradization

- **StandardScaler()** -> a preprocessing module used to standardize (or scale) data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(x)
```

```python
x_scaled = scaler.transform(x)
```

```python
x_scaled
```

# Regression with scaled features

```python
reg = LinearRegression()
reg.fit(x_scaled,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

# Creating a summary table

```python
reg_summary = pd.DataFrame([['Bias'],['SAT'],['Rand 1,2,3']], columns=['Features'])
reg_summary['Weights'] = reg.intercept_, reg.coef_[0], reg.coef_[1]
```

- The bigger the weight , the bigger the impact   
- The closer a weight is to 0 , the smaller its impact  

```python
reg_summary
```

# Making predictions with the standardized coefficients (weigts)

```python
new_data = pd.DataFrame(data=[[1700,2],[1800,1]],columns=['SAT','Rand 1,2,3'])
new_data
```

```python
reg.predict(new_data)
```

- The new data frame should be arranged in the same way..  and also must be standardizd in the same way

```python
new_data_scaled = scaler.transform(new_data)
new_data_scaled
```

```python
reg.predict(new_data_scaled)
```
