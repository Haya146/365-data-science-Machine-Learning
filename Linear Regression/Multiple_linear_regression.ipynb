## Load the data

```python
data = pd.read_csv('1.02. Multiple linear regression.csv')
data.head()
```

## Create multiple linear regression

### Declare dependent and independent variables

```python
x= data[['SAT','Rand 1,2,3']]
y= data['GPA']
```

### Regression itself

```python
reg = LinearRegression()
reg.fit(x,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

### Calculating R-squared

- The R-squared is a universal measure to evaluate how well linear regressions fare and compare   
- reg.score(x,y) -> returns the R-squared of a linear regression

```python
reg.score(x,y)
```

### Adjusted R-squared

- remember that : The adjusted R-squared is a much more appropriate measure for a multiple linear regression   
- The adjusted R-squared steps on the R-squared and adjusts for the number of variable included in the model   
- If we are using features with little or no explanatory power, the R-squared would increase, nonetheles. Thus, we need to penalize the successive usage through the adjusted R-squared

### **Formula for Adjusted $R^2$**
$R^2_{adj.} = 1 -(1 -R^2)*\frac{n-1}{n-p-1}$

- n =84 (the number of observations)  
- p =2 (the number of predictors )

```python
x.shape
```

```python
r2 = reg.score(x,y)
n = x.shape[0]
p = x.shape[1]
adjusted_r2 = 1-(1-r2)*(n-1)/(n-p-1)
adjusted_r2
```

# Feature selection

- How to detect variables whisch are unneeded in a model?   
- Feature selection is a very important procedure and machine learning as it simplifies models, which make them much easier to interpret by data scientist. Through this process.  
- Feature selection -> simplifies models, improves speed and prevents a series of unwanted issues arising from having too many features      
- in statsmodel we were saying that if a variable has a p-value > 0.05 we can disregard it   
- **f_refression** -> creates simple linear regressions of each feature and dependent variable
- **The p-value of F-stat = the p-value of the only independent variable**

```python
from sklearn.feature_selection import f_regression
```

```python
f_regression(x,y)
```

```python
p_values = f_regression(x,y)[1]
p_values
```

```python
p_values.round(3)
```
